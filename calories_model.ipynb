# Preprocessing the data

```python
#Change gender to dummy variable since it is a categorical data and find the square of duration and heart_rate to have more column
gender = pd.get_dummies(calories_df['Gender'], prefix = 'Gender',drop_first = True)
calories_df = pd.concat([calories_df, gender], axis = 1)
calories_df = calories_df.drop(['Gender'], axis=1)
calories_df['Duration^2'] = calories_df['Duration']**2
calories_df['Heart_Rate^2'] = calories_df['Heart_Rate']**2
calories_df['Body_Temp^2'] = calories_df['Body_Temp']**2
```

```python
calories_df.head()
```

```python
X = calories_df[['Age', 'Height', 'Weight', 'Gender_male','Duration','Heart_Rate', 'Body_Temp', 'Body_Temp^2', 'Duration^2', 'Heart_Rate^2']]
y = calories_df['Calories']
```

```python
#Normalize X, applying this to make the features be in the same range of 0-1
X_min = X.min()
X_max = X.max()
X = (X - X_min)/(X_max - X_min)
```

```python
"""
#Scale the numerical column with normal distribution z-score (standardization)
from scipy.stats import zscore
X = X.apply(zscore)

"""
```

```python
X.head()
```

```python
#Split the data into train test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

# Model Selection and Evaluation

#### The choice of using ploynomial regression was due to the relationship between the output feature (calories) and some of the input features

```python
# Covert the input features to polynomial features
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
```

```python
y_train_pred = poly_model.predict(X_train_poly)
poly_pred = poly_model.predict(X_test_poly)
```

```python
print('Intercept: ', poly_model.intercept_)
print('Coefficients: ', poly_model.coef_)
```

## Test Data Evaluation

```python
#creating a dataframe of the predicted value and the y_test
results = pd.DataFrame({'y_test': y_test, 'poly_pred': poly_pred})
results.head()
```

```python
#Regplot for the regression plot between the y_test and the predicted value
plt.figure(figsize = (12,6))
sns.regplot(x = 'y_test', y = 'poly_pred', data = results, line_kws={'color': 'red'})
plt.title('calories against predicted calories')
plt.xlabel('calories')
plt.ylabel('predicted calories')
```

```python
#Distplot to visualize the error in prediction 
plt.figure(figsize = (12,6))
sns.distplot((y_test - poly_pred), bins = 30)
plt.xlabel('error')
error = y_test - poly_pred
print('Maximum error:',error.max(),' Minimum error:', error.min())
```

```python
from sklearn import metrics
print('MAE: {:.2f}\nMSE: {:.2f}\nRMSE:{:.2f}\nR_score:{:.2f}'.format(metrics.mean_absolute_error(y_test, poly_pred),
                                                    metrics.mean_squared_error(y_test, poly_pred),
                                                    np.sqrt(metrics.mean_squared_error(y_test, poly_pred)),
                                                    metrics.r2_score(y_test, poly_pred)))
```

## Train data Evaluation

```python
#creating a dataframe of the predicted value and the y_train
results_train = pd.DataFrame({'y_train': y_train, 'y_train_pred': y_train_pred})
results_train.head()
```

```python
#Regplot for the regression plot between the y_train and the predicted value
plt.figure(figsize = (12,6))
sns.regplot(x = 'y_train', y = 'y_train_pred', data = results_train, line_kws={'color': 'red'})
plt.title('calories against train predicted calories')
plt.xlabel('calories')
plt.ylabel('train predicted calories')
```

```python
#Distplot to visualize the error in prediction 
plt.figure(figsize = (12,6))
sns.distplot((y_train - y_train_pred), bins = 30)
plt.xlabel('error')
error = y_train - y_train_pred
print('Maximum error:',error.max(),' Minimum error:', error.min())
```

```python
from sklearn import metrics
print('MAE: {:.2f}\nMSE: {:.2f}\nRMSE:{:.2f}\nR_score:{:.2f}'.format(metrics.mean_absolute_error(y_train, y_train_pred),
                                                    metrics.mean_squared_error(y_train, y_train_pred),
                                                    np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)),
                                                    metrics.r2_score(y_train, y_train_pred)))
```

#### The model gives 100% accuracy and have a mean absolute error of 1.95
